```python
# er_result.csv
def legend_without_duplicate_labels(ax):
    handles, labels = ax.get_legend_handles_labels()
    unique = [(h, l) for i, (h, l) in enumerate(zip(handles, labels)) if l not in labels[:i]]
    ax.legend(*zip(*unique))
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("er_result.csv")
df = df.sort_values(by=['數據集(Samples)', '數據集(Nodes)'], ascending=True)
df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,6))

# create data
x = np.arange(3)
y1 = df[(df['數據集(Samples)'] == 2000) & (df['演算法'] == 'Notears')]['used time']
y2 = df[(df['數據集(Samples)'] == 2000) & (df['演算法'] == 'Lingam')]['used time']
# y3 = df[(df['數據集(Samples)'] == 2000) & (df['演算法'] == 'GraNDAG')]['used time']
width = 0.2
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, y1, width, color='tab:blue')
# plt.bar(x, y2, width, color='tab:orange')
plt.bar(x+0.2, y2, width, color='tab:orange')
plt.xticks(x, ['20', '100', '300'])
plt.xlabel("Number of Nodes")
plt.ylabel("Used Time")
plt.legend(["Notears", "Lingam", "GraNDAG"])
plt.show()
```

### 誤報率 false discovery rate (FDR)

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,6))

target = "FDR"

# create data
x = np.arange(3)

y1 = df[(df['數據集(Samples)'] == 2000) & (df['演算法'] == 'Notears')][target]
y2 = df[(df['數據集(Samples)'] == 2000) & (df['演算法'] == 'Lingam')][target]
y3 = df[(df['數據集(Samples)'] == 2000) & (df['演算法'] == 'GraNDAG')][target]

# plot data in grouped manner of bar type
plt.plot(x, y1, color='tab:blue', label="Notears", marker="o", markerfacecolor='white')
plt.plot(x, y2, color='tab:orange', label="Lingam", marker="o", markerfacecolor='white')
# plt.plot(x, y3, color='tab:green', label="GraNDAG", marker="o", markerfacecolor='white')
plt.xticks(x, ['20', '100', '300'])

plt.xlabel("Number of Nodes")
plt.ylabel(target)

plt.legend()
legend_without_duplicate_labels(ax)
plt.show()
```
